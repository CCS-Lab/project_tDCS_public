```python
# Age, sex
```

```python
BRL_indices = pd.read_excel('/data2/project_BRL/project_tDCS/mri_session_checklist_data.xlsx')
```

```python
BRL_cov = BRL_indices.iloc[:,:3] 
```

```python
BRL_cov.columns = ['Session Name','male','age']
```

```python
sublist = []
for i in range(len(BRL_cov)):
    if i > 0:
        tmp = BRL_cov['Session Name'][i]
        tmp_ = tmp.split('_1')[0].split('BRL')[-1]
        sub = np.int(tmp_)
        sublist.append(sub)
```

```python
BRL_cov['sub'] = [0]+sublist
```

```python
BRL_cov = BRL_cov.replace('M',1).replace('F',0)
```

```python
BRL_cov.columns
```

```python
# Concat
```

```python
concat_ = pd.merge(BRL_cov, psych, on = 'sub', how = 'inner')
```

```python
concat = concat_.drop(['Session Name', 'No', 'SubjID'], axis = 1).set_index('sub')
```

```python
## N = 31
anode_first = [28, 18, 33, 60, 63, 64, 76, 91, 86, 59, 105, 107, 109, 112, 121, 126, 127]
sham_first = [15, 30, 54, 43, 11, 27, 73, 79, 101, 108, 110, 104, 125, 69]
```

```python
subjs = anode_first + sham_first
```

```python
concat = concat.loc[subjs] 
```

```python
anode = []
for i in range(len(concat)):
    if concat.index[i] in anode_first:
        anode.append(1)
    else:
        anode.append(0)
```

```python
concat['anode'] = anode
```
